# Handling Missing values:

```python
df=df.replace('?',np.nan)
df=df.replace('y',1)
df=df.replace('n',0)
```

# Splitting X and Y:

```python
X=df.drop('Class Name',axis='columns')
y=df['Class Name']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=0)
```

```python
print(len(X_train))
print(len(X_test))
```

# Modelling:

## Logistic Regression:

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(X_train,y_train)
y_pred=log.predict(X_test)
```

## Evaluation:

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

## Support Vector Classification:

```python
#SUPPORT VECTOR CLASSIFICATION: (Refer youtube SVM by Nachiketa Hebbar )
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
```

## Evaluation:

```python
print('Accuracy score of train data:',svc.score(X_train,y_train))
print('Accuracy score of test data:',svc.score(X_test,y_test))
```

# Hyper parameter tuning:

```python
#Hyper parameter tuning of SVC:
kernel=['linear','rbf','poly','sigmoid']
for i in kernel:
    svc=SVC(kernel=i,C=1.0)
    svc.fit(X_train,y_train)
    print('For Kernel:',i)
    print('Accuracy:',svc.score(X_test,y_test))
```

```python
for i in range(1,10):
    svc=SVC(kernel='linear',degree=i,C=1)
    svc.fit(X_train,y_train)
    print('Degree:',i)
    print('Accuracy of test data:',svc.score(X_test,y_test))
```

# Gridsearch CV:

## SVC:

```python
#Final hyperparameter tuning using GRIDSEARCHCV:(Finding all the best hyperparameter using one single library 'GridSearchCV')
from sklearn.model_selection import GridSearchCV

param_grid={'C':[0.1,1,100,1000],'kernel':['rbf','linear','poly','sigmoid'],'degree':[1,2,3,4,5]}
grid=GridSearchCV(SVC(),param_grid)
grid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
print(grid.score(X_test,y_test))
```
